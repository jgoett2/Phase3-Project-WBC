```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

print(cv.__version__)
```

```python
img = cv.imread("Data/RawImgs/BloodImage_00001.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img = np.array(img)
img.shape

plt.imshow(img)
```

```python
def mask(img):
    for row in img:
      for pixel in row:
        if ((pixel[0]<166) and (pixel[1]<153) and (pixel[2]>152)):
          pixel[0] = 255
          pixel[1] = 255
          pixel[2] = 255
        else:
          pixel[0] = 0
          pixel[1] = 0
          pixel[2] = 0
    return img
```

```python
img2 = mask(img)
plt.imshow(img2)
```

```python
img3 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY) 
contours, _ = cv.findContours(img3, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

# Find the largest contour
largest_contour = max(contours, key=cv.contourArea)

cv.drawContours(img3, [largest_contour], -1, (0, 255, 0), 3)

# Display the image
plt.imshow( img3)
```

```python
temp = largest_contour[:,0,0]
x_min = min(temp) - 10
x_max = max(temp) + 10

temp = largest_contour[:,0,1]
y_min = min(temp) - 10
y_max = max(temp) + 10
```

```python
def center_mass(img):
    total = 0
    weighted_total = 0
    for y in range(img.shape[0]):
        temp = img[y].sum()
        total += temp
        weighted_total += y*temp
    y_com = weighted_total/total
    total = 0
    weighted_total = 0
    for x in range(img.shape[1]):
        temp = img[:,x].sum()
        total += temp
        weighted_total += x*temp
    x_com = weighted_total/total
    return(x_com, y_com)
        
def center_image(img):
    x_com, y_com = center_mass(img)
    tx = 320 - x_com
    ty = 240 - y_com
    M = np.float32([[1, 0, tx], [0, 1, ty]])
    shifted_image = cv.warpAffine(img, M, (640, 480))
    return shifted_image
```

```python
img3 = img2
shifted_image = center_image(img3)
plt.imshow(shifted_image[140:340,220:420])
```

## Crop Entire Library

```python
import os
X = []
for filename in os.listdir("Data/RawImgs/"):
    if filename[:5] == "Blood":
      img = cv.imread("Data/RawImgs/" + filename )
      img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
      img = mask(img)
      shifted_image = center_image(img)[140:340,220:420]
      X.append({"file": filename, "data": shifted_image})
      cv.imwrite("Data/BWImgs/"+ filename, shifted_image)
      
```

```python
df = pd.read_csv("Data/WBC_Labels.csv")
df["FILENAME"] = df["FILENAME"].str.strip("'")
df = df.set_index("FILENAME")
```

```python
X_transform = []
for data in X:
    temp = data["data"][:,:,0].reshape(-1)
    temp2 = data["file"]
    X_transform.append({"file":temp2, "data":temp})
```

```python
df2 = pd.DataFrame(X_transform).set_index("file")
df2
```

## Try a quick KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.concat([df, df2],axis=1)
df
```

```python
X = df["data"].to_numpy()
X = pd.DataFrame(df["data"].to_list())
```

```python
y = df["FAMILY"]
```

```python
X_train = X[:250]
X_test = X[250:]
y_train = y[:250]
y_test = y[250:]
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
print(knn.score(X_test, y_test))
y_test_predict = knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_test_predict)
```
